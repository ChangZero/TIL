```python
def Duggi():
    storelst = []
    addresslst = []
    numberlst = []

    for i in range(1,31):
        url = f'https://twozzim.com/bbs/board.php?bo_table=store&page={i}'
        html = urllib.request.urlopen(url)
        soup = BeautifulSoup(html, 'lxml')

        tbl = soup.find('div', attrs={'class': 'tbl_head01 tbl_wrap'})

        tabl_tds = tbl.find_all('ul', class_='tabl_td')

        regst = re.compile(r'>([가-힣|a-z|A-Z|0-9]+)</a>')
        regad = re.compile(r'>([가-힣|a-z|A-Z|0-9]+)</li>')
        regnum = re.compile(r'>([0-9]+-[0-9]+-[0-9]+)</a>')

        for td in tabl_tds:
            li1 = td.find('li', attrs={'class':'tabl_th_li1 font-r'}).text
            li2 = td.find('li', attrs={'class':'tabl_th_li2 font-r'})
            li3 = td.find('li', attrs={'class':'tabl_th_li3_left font-r'}).text

            number = re.findall(regnum, str(li2))
            storelst.append(li1)
            addresslst.append(li3)
            numberlst.append(str(number)[2:-2])
            
    return storelst, addresslst, numberlst
```

```python
a,b,c = Duggi()
```
